## Ex 5

```python

df = pd.read_csv("ex4.csv")
df = df.sort_values(by="degree",ascending=False)

hist = Counter(df.degree)

fig = plt.figure()
ax = plt.gca()
ax.set_yscale('log')

ax.scatter(hist.keys(),hist.values())

plt.xlabel("degree")
plt.ylabel("number of vertices")


plt.show() # tak sieć jest bezskalowa. Wiele pojedynczych nodów jest przyczepionych do dużych hubów
```
